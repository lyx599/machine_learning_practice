## 客观探索

````python
corr_matrix = train_set.drop(['Id'],axis=1).corr()
fig,ax = plt.subplots(figsize=(15,10))
sns.heatmap(corr_matrix, vmax=0.8, square=True,center=0,cmap='RdBu_r');
````

从最后一行可以看到与SalePrice相关比较大的有：OverallQual，TotalBsmtSF，1stFlrSF，GrLivArea，GarageCars，GarageAre  
稍弱一点的有：YearBuilt,YearRemodAdd,MasVnrArea,FullBath,TotRmsAbvGrd,GarageYrBlt

其中TotalBsmtSF与1stFlrSF，GarageCars与GarageAre，YearBuilt与GarageYrBlt，TotRmsAbvGrd与GrLivArea有极强的相关性（选取的颜色为深红色的，其他的未考虑），这几个特征需在后面进行筛选

相关系数都是数值型的，没有对类别型的数据的分析

````python
#选出与SalePrice前10相关性的特征
top = 11
cols = corr_matrix.nlargest(top,'SalePrice')['SalePrice'].index
cm = np.corrcoef(train_set[cols].values.T)#corrcoef相关系数
plt.figure(figsize=(12,10))
sns.heatmap(cm,cbar=True,annot=True,square=True,fmt='.2f',annot_kws={'size':10},
           yticklabels=cols.values,xticklabels=cols.values)
plt.show()
````

````python
corr_matrix.nlargest(top,'SalePrice')['SalePrice']
````

通过编程进行准确量化特征，和之前根据上个热力图做出的判断差不多，因为存在共线特征，所以增加范围选出top10数值特征

## 缺失值处理

````python
train_set[num_attributes].info()
````

缺失值较少，直接使用中位数填补

````python
train_set[nonum_attributes].info()
````

````python
from sklearn.preprocessing import Imputer
imputer = Imputer(strategy='median')
imputer.fit(train_set[num_attributes])
````

````python
X_num = imputer.transform(train_set[num_attributes])
X_num.shape
````

````python
train_set[['MSZoning','BsmtQual']].groupby(['MSZoning','BsmtQual']).size()
````

````python
for ID in train_set[train_set['BsmtQual'].isna()].index:
    if train_set.loc[ID,'MSZoning'] == 'C':
        train_set.loc[ID,'BsmtQual'] = 'Ex'
    elif train_set.loc[ID,'MSZoning'] == 'FV':
        train_set.loc[ID,'BsmtQual'] = 'Gd'
    elif train_set.loc[ID,'MSZoning'] == 'RH':
        train_set.loc[ID,'BsmtQual'] = 'TA'
    elif train_set.loc[ID,'MSZoning'] == 'RL':
        train_set.loc[ID,'BsmtQual'] = 'Gd'
    elif train_set.loc[ID,'MSZoning'] == 'RM':
        train_set.loc[ID,'BsmtQual'] = 'TA'   
````

````python
train_set[train_set['BsmtQual'].isna()]
````

````python
train_set[nonum_attributes].info()
````

## 文本特征转数字特征

````python
train_set[nonum_attributes].head()
````

````python
nonum_attributes_tr = pd.get_dummies(train_set[nonum_attributes])
nonum_attributes_tr.head()
````

## 正态性

````python
from scipy.stats import norm
from scipy import stats
sns.distplot(train_set['SalePrice'],fit=norm)
fig = plt.figure()
res = stats.probplot(train_set['SalePrice'],plot=plt)
````

可以看出，房价分布不是正态的，显示了峰值，正偏度，但是并不跟随对角线。

````python
train_set['SalePrice'] = np.log(train_set['SalePrice'])
````

````python
sns.distplot(train_set['SalePrice'],fit=norm)
fig = plt.figure()
res = stats.probplot(train_set['SalePrice'],plot=plt)
````

## 特征缩放

````python
#使用这个缩放可以更好的处理异常值
from sklearn.preprocessing import RobustScaler
std = RobustScaler()
X_num_std = std.fit_transform(X_num)
````

````python
X_num_std.shape
````

## 完成数据 

````python
features = pd.concat([train_set[num_attributes],nonum_attributes_tr],axis=1)
features.head()
````

````python
X_nonum = nonum_attributes_tr.values
X_nonum.shape
````

````python
X = np.c_[X_num_std,X_nonum]
X.shape
````

````python
y = train_set['SalePrice']
y.shape
````

# 特征筛选

````python
from sklearn.ensemble import RandomForestRegressor

rnd_reg = RandomForestRegressor(n_estimators=50,max_features='sqrt')
rnd_reg.fit(X,y)
````

````python
features_imp = pd.DataFrame()
features_imp['feature'] = features.columns
features_imp['importance'] = rnd_reg.feature_importances_
features_imp.sort_values(by=['importance'], ascending=True, inplace=True)
features_imp.set_index('feature', inplace=True)
features_imp.plot(kind='barh',figsize=(12,10))
````

````python
from sklearn.feature_selection import SelectFromModel
model = SelectFromModel(rnd_reg, prefit=True)
X_train_reduced = model.transform(X)
X_train_reduced.shape
````

可以使用sklearn自带的库进行选择，也可以直接根据需要选择

# 不同模型进行测试

## 线性模型

````python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train_reduced,y)
````

````python
from sklearn.metrics import mean_squared_error
y_lin_reg_pred = lin_reg.predict(X_train_reduced)
lin_mse = mean_squared_error(y_lin_reg_pred,y) 
lin_rmse = np.sqrt(lin_mse)
lin_rmse
````

````python
from sklearn.model_selection import cross_val_score

np.sqrt(-cross_val_score(lin_reg, X_train_reduced, y, cv=3, scoring='neg_mean_squared_error'))
````

## 随机森林

````python
from sklearn.ensemble import RandomForestRegressor

rnd_reg = RandomForestRegressor()
rnd_reg.fit(X_train_reduced,y)
````

````python
y_rnd_reg_pred = rnd_reg.predict(X_train_reduced)
rnd_mse = mean_squared_error(y_rnd_reg_pred,y) 
rnd_rmse = np.sqrt(rnd_mse)
rnd_rmse
````

````python
from sklearn.model_selection import cross_val_score

cross_val_score(rnd_reg, X_train_reduced, y, cv=3, scoring='neg_mean_squared_error')
````

这里要注意的是，上面的scores都是负数，为什么均方误差会出现负数的情况呢？因为sklearn里 ```greater_is_better ``` 
所以均方误差前面有个neg

````python
from sklearn.model_selection import cross_val_score

np.sqrt(-cross_val_score(rnd_reg, X_train_reduced, y, cv=3, scoring='neg_mean_squared_error'))
````

这里值就表示越大效果越差,随机森林过拟合

## 模型微调

````python
from sklearn.model_selection import GridSearchCV
parameter_grid = {
                 'max_depth' : [2, 3, 4, 5, 6, 7, 8],
                 'n_estimators': [10,20,30,40,50],
                 'max_features': ['sqrt', 'auto', 'log2'],
                 'min_samples_split': [2, 3, 5, 10],
                 'min_samples_leaf': [1, 3, 5, 10],
                 'bootstrap': [True, False],
                 }
forest = RandomForestRegressor()

grid_search = GridSearchCV(forest,
                            scoring='neg_mean_squared_error',
                            param_grid=parameter_grid,
                            cv=3,
                            )

grid_search.fit(X_train_reduced, y)
model = grid_search
parameters = grid_search.best_params_
````

````python
parameters
````

````python
model
````

````python
np.sqrt(-cross_val_score(model, X_train_reduced, y, cv=3, scoring='neg_mean_squared_error'))
````

````python
rnd_model = grid_search.best_estimator_ 
y_model_pred = rnd_model.predict(X_train_reduced)
model_mse = mean_squared_error(y_model_pred,y) 
model_rmse = np.sqrt(model_mse)
model_rmse
````

````python
np.sqrt(-cross_val_score(rnd_model, X_train_reduced, y, cv=3, scoring='neg_mean_squared_error'))
````
